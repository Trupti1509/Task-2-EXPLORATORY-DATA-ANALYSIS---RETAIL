## Analysis using Pairplot of column

```python
# Category column analysis
sns.pairplot(df,hue='Category')
```

# Heatmap for Correlation

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
```

From the above heatmap:
*  Sales and profit are moderately corelated
*  Discount and Profit are Negatively Correlated
* Quantity and Profit are less Moderately Correlated








# Barplot of the column

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Profit', data=df)
plt.title("Profits by Category")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Profit', data=df)
plt.title("Profits by Region")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Segment', y='Profit', data=df)
plt.title("Profits by Customer Segment")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Ship Mode', y='Profit', data=df)
plt.title("Profits by Ship Mode")
plt.show()
```

# Scatter plot of Profits vs. Discount

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title("Profits vs. Discount")
plt.show()
```

## What are the business problems you can derieve by exploring the data?

Based on the provided Google Colab code and analysis, here are some key insights and potential business problems that a business manager can derive from exploring the data:

1. *Profit by Category:*
   - The "Furniture" category appears to have the lowest profitability. A business manager should investigate why furniture is less profitable and consider strategies to improve it.

2. *Profit by Region:*
   - The Central region has the lowest profit. The business manager should analyze the factors contributing to this lower profitability, such as pricing, competition, or cost efficiency, and develop strategies to address them.

3. *Profit by Customer Segment:*
   - The "Home Office" segment has the lowest profitability. The business manager can investigate the reasons behind this and determine whether changes in marketing, pricing, or product offerings can improve profitability for this segment.

4. *Profit by Ship Mode:*
   - Standard Class shipping mode appears to have lower profitability compared to other modes. The business manager should assess why this is the case and consider changes in shipping strategies or costs.

5. *Discount vs. Profit:*
   - The scatterplot shows a negative correlation between discounts and profit. Business problems related to excessive discounting might be addressed to improve overall profitability.

6. *Correlation Analysis:*
   - The correlation heatmap indicates the relationships between various numerical variables. A business manager can identify areas where variables are strongly correlated (positively or negatively) and take action accordingly. For example, a high negative correlation between "Discount" and "Profit" suggests that discounting may be affecting profit negatively.

7. *Sales by Category:*
   - While not explicitly mentioned in the code, analyzing sales by category can reveal which product categories are the most and least popular. This information can help in making informed decisions about inventory and marketing strategies.

8. *Location-Based Analysis:*
   - Analyzing profits by city and state can help identify specific locations with lower profitability. Business managers can then investigate these locations to understand the root causes and improve profitability.

9. *Product Sub-Category Analysis:*
   - Further analysis of profits by product sub-categories can identify specific product lines or items that are underperforming. The business manager can make decisions regarding pricing, marketing, or discontinuation of these products.

10. *Time-Series Analysis (if date data is available):*
    - If the dataset contains date information, a business manager can perform time-series analysis to identify seasonal trends and patterns that can inform inventory management and marketing strategies.
